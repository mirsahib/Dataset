```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('full_position.csv')
```

```python
fromRow = 0
toRow = 250
#accelerometer
x = df.iloc[fromRow:toRow,1]
y = df.iloc[fromRow:toRow,2]
z = df.iloc[fromRow:toRow,3]
plt.plot(x,label='x')
plt.plot(y,label='y')
plt.plot(z,label='z')
plt.legend()
plt.title("Accelerometer")
plt.grid()
plt.show()
#gyroscope
x = df.iloc[fromRow:toRow,4]
y = df.iloc[fromRow:toRow,5]
z = df.iloc[fromRow:toRow,6]
plt.plot(x,label='x')
plt.plot(y,label='y')
plt.plot(z,label='z')
plt.legend()
plt.title("Gyroscope")
plt.grid()
plt.show()
#Magnetometer
x = df.iloc[fromRow:toRow,7]
y = df.iloc[fromRow:toRow,8]
z = df.iloc[fromRow:toRow,9]
plt.plot(x,label='x')
plt.plot(y,label='y')
plt.plot(z,label='z')
plt.legend()
plt.title("Magnetometer")
plt.grid()
plt.show()
```
