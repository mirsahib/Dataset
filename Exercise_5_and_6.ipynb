```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew,kurtosis,iqr
from statsmodels import robust
```

```python
df = pd.read_csv('IMU_Mega.csv')
```

```python
fromRow = 5880
toRow = 6401
#accelerometer
x = df.iloc[fromRow:toRow,1]
y = df.iloc[fromRow:toRow,2]
z = df.iloc[fromRow:toRow,3]
plt.plot(x,label='x')
plt.plot(y,label='y')
plt.plot(z,label='z')
plt.grid()

plt.legend()
plt.title("Accelerometer")
plt.show()
#gyroscope
x = df.iloc[fromRow:toRow,4]
y = df.iloc[fromRow:toRow,5]
z = df.iloc[fromRow:toRow,6]
plt.plot(x,label='x')
plt.plot(y,label='y')
plt.plot(z,label='z')
plt.legend()
plt.title("Gyroscope")
plt.show()
#Magnetometer
x = df.iloc[fromRow:toRow,7]
y = df.iloc[fromRow:toRow,8]
z = df.iloc[fromRow:toRow,9]
plt.plot(x,label='x')
plt.plot(y,label='y')
plt.plot(z,label='z')
plt.legend()
plt.title("Magnetometer")
plt.show()
```

```python
nintyDegree = []
zeroDegree = []
movement = []
for i in range(fromRow,toRow):
    if df.iloc[i][1]<=9 and df.iloc[i][1]>=5.5:
        nintyDegree.append(i)
    elif df.iloc[i][1]<=-4.5 and df.iloc[i][1]>=-5.5:
        zeroDegree.append(i)
    elif df.iloc[i][1]<8.5 and df.iloc[i][1]>-5.2:
        movement.append(i)
```
